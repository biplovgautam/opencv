```python
import cv2
import matplotlib.pyplot as plt

# Filenames for the 3 images
image_files = ['1.jpeg', '2.jpeg', '3.jpeg']

images = []
for file in image_files:
    img = cv2.imread(file)
    if img is None:
        print(f"Error loading {file}. Please check the filename and path.")
        continue
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    images.append(img)

print(f"Loaded {len(images)} images.")
```

```python
plt.figure(figsize=(15, 5))
for i, img in enumerate(images):
    plt.subplot(1, 3, i+1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'Image {i+1}')
plt.show()
```

```python
stitcher = cv2.Stitcher_create()
status, panorama = stitcher.stitch(images)

if status == cv2.Stitcher_OK:
    plt.figure(figsize=(20, 10))
    plt.imshow(panorama)
    plt.axis('off')
    plt.title("Panorama Result")
    plt.show()

    # Save the panorama image (convert RGB back to BGR for saving)
    cv2.imwrite('stitchedpanorama.jpeg', cv2.cvtColor(panorama, cv2.COLOR_RGB2BGR))
    print("Panorama saved as 'stitchedpanorama.jpeg'")
else:
    print(f"Stitching failed with error code {status}")
```
